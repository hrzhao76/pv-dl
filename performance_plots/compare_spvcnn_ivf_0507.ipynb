# Compare reco HS eff

```python
def convert_TEfficiency_to_array(teff):
    """Convert ROOT.TEfficiency to numpy array."""
    tgr = teff.CreateGraph()

    bin_properties = []
    for i in range(gr.GetN()):
        bin_properties.append([tgr.GetX()[i], tgr.GetY()[i], tgr.GetEYlow()[i], tgr.GetEYhigh()[i]])

    bin_properties = np.array(bin_properties)
    x = bin_properties[:, 0]
    y = bin_properties[:, 1]
    yerr = np.vstack((bin_properties[:, 2], bin_properties[:, 3]))

    return x, y, yerr
```

```python
fig, ax = plt.subplots()
for itr_idx, itr in enumerate(itrs_results):
    x, y, yerr = convert_TEfficiency_to_array(itr[2])
    ax.errorbar(
        x, y, yerr=yerr, fmt=markers_map[itr_idx], label=labels_map[itr_idx], color=colors_map[itr_idx]
    )
ax.set_xlim(0, 2)
ax.legend()
ax.set_title("HS Reconstruction Efficiency")
ax.set_xlabel("Local PU density")
ax.set_ylabel("Efficiency")

fig.show()
fig.savefig(output + "hs_reco_efficiency_compare.png", dpi=300)
```
